```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
x = np.arange(0, 9, 0.1)
y = np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
x = np.arange(0, 9, 0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.plot(x,y_sin)
plt.plot(x,y_cos)
plt.xlabel('X-axis label')
plt.ylabel('Y-axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine','Cosine'])

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
x = np.arange(0, 9, 0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)
plt.plot(x,y_sin)
plt.title('Sine')
plt.subplot(2,1,2)
plt.plot(x,y_cos)
plt.title('Cosine')
plt.show()
```
